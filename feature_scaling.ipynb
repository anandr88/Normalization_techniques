<h1 style="text-align: justify; font-family: 'WildWest'; font-size:25px; color:IndianRed"> Feature Scaling </h1>

Feature Scaling is a technique to standardize the independent features present in the data in a fixed range.

<div class="alert alert-block alert-info" style="
                                                 color:black;
                                                 font-size:120%;
                                                 border-color:black;
                                                 border-radius:5px;
                                                 width:100%;
                                                 margin: auto;
                                                 text-align: left;">
<li>Normalization is the process of changing the values of numbers in a dataset to a common scale. 
<br><br><li>This helps in making sure that no single number or set of numbers dominates the others, while keeping the original differences between the numbers the same.
<br><br><li>The terms normalization and standardization are sometimes used interchangeably, but they usually refer to different things.
<br><br><li>The main purpose is to bring all feature values into a similar range to improve the performance and training stability of machine learning models.
<br><br><li> Normalization (using MinMaxScaler) is always in a range of [0,1] or [-1,1] if there are negative values. 
<br>
<br>
<li>Standardization (or Z-score normalization) is that the features will be rescaled to ensure the mean and the standard deviation to be 0 and 1, respectively. Standardization is not bounded by range;
</div>
<center><img src="https://www.someka.net/wp-content/uploads/2020/08/difference-between-normalization-standardization-1.png" alt="Heart Disease" width="60%"></center>

### Benefits of Normalization:

**Improves Convergence:** Helps gradient descent converge faster in training machine learning models.

**Stability:** Provides numerical stability in algorithms, preventing overflow or underflow.

**Equal Weight:** Ensures that all features contribute equally to the model, preventing dominance by features with larger scales.

**Interpretability:** Makes coefficients in linear models more interpretable.

<div class="alert alert-block alert-info" style="
                                                 color:black;
                                                 font-size:120%;
                                                 border-color:black;
                                                 border-radius:5px;
                                                 width:100%;
                                                 margin: auto;
                                                 text-align: left;">
<li>Standardization is preferred for algorithms that are sensitive to feature scale (Gradient Descent Based Algorithms) or assume normality, such as linear regression, logistic regression, neural network, PCA (principal component analysis), etc., 
    <br>
<br>
<li>while normalization is better suited for distance-based algorithms like KNN, K-means clustering, and SVM(support vector machines).
    
</div>

## <b><span style='color:#E1B12D'> Is it valid to standardize variables with non-normal distribution?</span></b>

Once again it is worth highlighting: StandardScaler will not affect data distribution. If you thought it was transforming the distribution of the data, you need to get out of that concept. 

It's important to keep in mind that StandardScaler only standardizes the features in the data, it does not enforce a normal distribution.

.
<div style=" background-color:#3b3745;
            text-align:left;
            padding: 13px 13px; 
            border-radius: 8px; 
            margin: auto;
            color: white">
<ul>
<li>There are no possible issues with standardizing non-normal distribution. In other words, if the features do not follow a normal distribution before standardization, they will not follow it after the standardization either (the distribution will remain exactly the same - only the scale will change). So, it‚Äôs valid to do it on a non-normal distribution.
 <br>
<br>   
<li>Standardization also does not assume anything about the distribution of variables (normal or other).
<br>
<br>
<li>Go ahead and standardize the variables (if required) without worrying about the distribution!
    <br>
<br>
<li>But be careful when a data is highly skewed or have significant outliers (topic discussed later)!

</div>


<h1 style="text-align: justify; font-family: 'WildWest'; font-size:30px; color:IndianRed">Mean-Normalization</h1>


<div style=" background-color:#3b3745;
            text-align:left;
            font-size:130%;
            padding: 13px 13px; 
            border-radius: 8px; 
            margin: auto;
            color: white">
<ul>
    <li>Mean normalization is very similar to min-max normalization. 
        <br>
<br>

<li>The difference is instead of the minimum value, the mean value of the column is subtracted in the numerator. The denominator is the same, i.e. the difference between the maximum and the minimum value.
<br>
<br>
Mean-Normalization formula:

$x' =  \frac {x ‚àí Œº}{max(x)-min(x)}$

mean = Œº
    
<br>
<br><li>In other words, the mean value of a numerical column is subtracted from each value of the column. And then, the result is used as a numerator. 

<br>
<br><li>Then, the minimum value of the column is subtracted from the maximum value of the column. And the result is used as a denominator. 

<br>
<br><li>We divide the numerator by the denominator and the result is used to replace the value of the column.

```python
df = pd.read_csv("titanic.csv", usecols=["Age", "Fare"])
df["Fare"].fillna(value=df["Fare"].mean(), inplace=True)
```

```python
def mean_normalization(df):
    mean = df.mean()
    min_val = df.min()
    max_val = df.max()
    df_normalized = (df - mean) / (max_val - min_val)
    return df_normalized

# Apply mean normalization
df_new_mm = mean_normalization(df)

# plotting the scatterplot of before and after Min Max Scaling
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title("Scatterplot Before mean normalization Scaling", fontsize=18)
sns.scatterplot(data = df, color="blue")
plt.subplot(1,2,2)
plt.title("Scatterplot After mean normalization Scaling", fontsize=18)
sns.scatterplot(data = df_new_mm, color="red")
plt.tight_layout()
plt.show()
```

```python

# plotting the scatterplot of before and after Min Max Scaling
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title(" PDF Before mean normalization Scaling", fontsize=18)
sns.kdeplot(data = df, color="blue")
plt.subplot(1,2,2)
plt.title("PDF After mean normalization Scaling", fontsize=18)
sns.kdeplot(data = df_new_mm, color="red")
plt.tight_layout()
plt.show()
```

<ul>
    Scaling using median and quantiles consists of substracting the median to all the observations, and then dividing by the interquantile difference. The interquantile difference is the difference between the 75th and 25th quantile:
<br>
<br>
IQR = 75th quantile - 25th quantile
    <br>
    <br>
<br>
RobustScaler formula:

$x' =  \frac {x ‚àí median(x)}{IQR}$
<br>
<br>
IQR - describes the middle 50% of values when ordered from lowest to highest (IQR = Q3 ‚Äì Q1).
<br>
<br>
The scaled values will have their median and IQR set to 0 and 1, respectively.
</div>

    <li>Different approach to standardizing input variables in the presence of outliers is to ignore them from the calculation of the mean and standard deviation, then use the calculated values to scale the variable.

<br><br><li>Unlike the previous scalers, the centering and scaling statistics of RobustScaler are based on percentiles and are therefore not influenced by a small number of very large marginal outliers. 
    <br><br><li>Consequently, the resulting range of the transformed feature values is larger than for the previous scalers and, more importantly, are approximately similar.

<br><br><li>The resulting variable is not skewed by outliers and the outliers are still present with the same relative relationships to other values.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# Creating functions for scaling
def Standard_Scaler (df, col_names):
    features = df[col_names]
    scaler = StandardScaler().fit(features.values)
    features = scaler.transform(features.values)
    df[col_names] = features
    return df

def MinMax_Scaler (df, col_names):
    features = df[col_names]
    scaler = MinMaxScaler().fit(features.values)
    features = scaler.transform(features.values)
    df[col_names] = features
    return df

def Robust_Scaler (df, col_names):
    features = df[col_names]
    scaler = RobustScaler().fit(features.values)
    features = scaler.transform(features.values)
    df[col_names] = features
    return df
```

It is important to remember that the fit method is used for calculating the mean and variance of each of the features present in our data set. The transform method is transforming all the features using the respective mean and variance. In this cose we transform the entire data set without splitting, but when we perform train-test split we have to remember that fit and transform is for train and transform for test set.

## <b> <span style='color:#E1B12D'>Is it always better to use the RobustScaler?</span></b>

When analyzing data, it's important to consider the amount of information that can be extracted from it. Robust measures like the interquartile range (IQR) are designed to focus on a specific subset of the data while ignoring potentially unreliable values. However, this deliberate exclusion of certain data can limit the overall amount of information that is used to make predictions, leading to lower confidence of the result.

## <b><span style='color:#E1B12D'>Best practices</span></b>

<div style=" background-color:#3b3745;
            text-align:left;
            padding: 13px 13px; 
            border-radius: 8px; 
            margin: auto;
            color: white">
<ul> 
<li> It‚Äôs usually a bad idea to train machine learning models on features with vastly different scales.
<li> Consider deleting outliers before scaling a data, but be very careful when a data is highly imbalanced.
<li> Some machine learning algorithms are sensitive to feature scaling (gradient descent, distance based) while others are not (tree-based).
<li> Normalize or standardize always AFTER splitting data (of course you need to apply normalization/standardization to test data as well).
<li> StandardScaler, MinMaxScaler and RobustScaler will not affect data distribution (the algorithms don't enforce a normal distribution).

<li> Normalisation does not treat outliers well (MinMax Scaler shrinks the data within the given range which will affect the ability of the algorithm to give adequate weights for the features).
<li> MinMaxScaler can be used when the upper and lower boundaries are well known from domain knowledge.

<li> Standardization allows better handling of the outliers and better facilitates convergence of features (which is important for the gradient descent and distance based algorithms).
<li> StandardScaler is also sensitive to outliers (less than MinMaxScaler, but still sensitive). The mean and standard deviation are highly affected by outliers so the distance (measured in standard deviation) is also affected by outliers - the algorithm will scale most of the data to a small interval.
<li> RobustScaler's results are not skewed by outliers and the spread represents REAL distance.
<li> RobustScaler's deliberate exclusion of certain data can limit the overall amount of information that is used to make predictions, leading to lower confidence of the result.
<li> In my opinion there are no possible issues with standardizing non-normal distribution. 

<li> You can always start by fitting your model to raw, normalized and standardized data and compare the performance for best results.

***

* I decide to delete outliers or they are not present in a data set;
* when the data is not very highly skewed.

I would use RobustScaler in opposite situations and MinMaxScaler only in a very special cases.

```python
Q1 = df.quantile(0.15)
Q3 = df.quantile(0.85)
IQR = Q3 - Q1
IQR.sort_values(ascending=False)
```

```python
wine1 = df [~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
X = wine1.drop(['quality'],axis=1) 
y = wine1['quality']
```

There are a few straightforward methods to transform variables so that they follow a Gaussian distribution. None of them is better than the other, they rather depend on the original distribution of the variables. They are:

    * a. Logarithmic Transformation
    * b. Reciprocal Transformation
    * c. Square Root Transformation
    * d. Exponential Transformation
    * e. Box Cox Transformation

### Fill missing data with random sample

```python
# first I will fill the missing data of the variable age, with a random sample of the variable

def impute_na(data, variable):
    # function to fill na with a random sample
    df = data.copy()
    
    # random sampling
    df[variable+'_random'] = df[variable]
    
    # extract the random sample to fill the na
    random_sample = df[variable].dropna().sample(df[variable].isnull().sum(), random_state=0)
    
    # pandas needs to have the same index in order to merge datasets
    random_sample.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(), variable+'_random'] = random_sample
    
    return df[variable+'_random']
```

```python
# fill na
data['Age'] = impute_na(data, 'Age')
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab
```

```python
# plot the histograms to have a quick look at the distributions
# we can plot Q-Q plots to visualise if the variable is normally distributed

def diagnostic_plots(df, variable):
    # function to plot a histogram and a Q-Q plot
    # side by side, for a certain variable
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    df[variable].hist()

    plt.subplot(1, 2, 2)
    stats.probplot(df[variable], dist="norm", plot=pylab)

    plt.show()
    
diagnostic_plots(data, 'Age')
```

<li>BoxCox transformation

<li>The Box-Cox transformation is defined as:

T(Y)=(Y exp(Œª)‚àí1)/Œª

<li>where Y is the response variable and Œª is the transformation parameter. Œª varies from -5 to 5. In the transformation, all values of Œª are considered and the optimal value for a given variable is selected.

<li>Briefly, for each Œª (the transformation tests several Œªs), the correlation coefficient of the Probability Plot (Q-Q plot below, correlation between ordered values and theoretical quantiles) is calculated.

<li>The value of Œª corresponding to the maximum correlation on the plot is then the optimal choice for Œª.

<li>In python, we can evaluate and obtain the best Œª with the stats.boxcox function from the package scipy.

Let's have a look.

```python
data['Age_boxcox'], param = stats.boxcox(data.Age) 

print('Optimal Œª: ', param)

diagnostic_plots(data, 'Age_boxcox')
```
